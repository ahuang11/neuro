```python
import numpy as np
import holoviews as hv; hv.extension('bokeh')
```

```python
curves = hv.Overlay([hv.Curve(np.random.rand(100), label=str(i)).opts(
    subcoordinate_y=True) for i in np.arange(10)]).opts(
    show_legend=False, ylim=(0,1))
```

```python
hv.Image(np.random.rand(100, 100), bounds=(0,0,10,10)).opts(
    xlim=(0, 1), ylim=(0, 1),  # initial display extent
    backend_opts={"x_range.max_interval":0.5},
    default_tools=['zoom_in', 'zoom_out', 'reset', ])
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import Range1d
from bokeh.io import output_notebook

output_notebook()

p = figure(tools=['zoom_in', 'zoom_out', 'reset'],
           x_range=Range1d(start=0, end=50, max_interval=25))

p.image(image=[np.random.rand(50, 50)], x=0, y=0, dw=50, dh=50)

# Set initial display limits
p.x_range.start = 10
p.x_range.end = 35
p.y_range.start = 10
p.y_range.end = 35

show(p)
```
