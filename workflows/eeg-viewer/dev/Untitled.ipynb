```python
# Complete working example code for setting max_interval in a Bokeh time-series plot

from bokeh.plotting import figure, show, output_notebook
from datetime import datetime, timedelta
import numpy as np

# Initialize Bokeh notebook output
output_notebook()

# Generate example time-series data
n = 100
base = datetime.today()
date_list = [base - timedelta(days=x) for x in range(n)]
y_values = np.random.rand(n) * 100

# Create the figure
p = figure(width=800, height=400, x_axis_type="datetime", title="Time Series with Max Zoom-Out Limit")

# Add a line renderer
p.line(date_list, y_values, line_width=2)

# Set max_interval on x-axis to limit zoom-out extent to 30 days (in milliseconds)
p.x_range.max_interval = 30 * 24 * 60 * 60 * 1000  # 30 days in milliseconds

# Show the plot
show(p)
```
